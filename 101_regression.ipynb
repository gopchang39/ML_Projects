### Call library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Univariate Regression
### Dummy data

```python
# sample data
X = np.array([1,2,3,4])
y = np.array([2,1,4,3])
```

### Plot data

```python
# plot data: x vs y
plt.scatter(X, y)
```

### Data preprocessing

```python
# data conversion for scikit-learn
# form of (n_samples, n_features)
X.shape
```

```python
data = X.reshape(-1, 1)
data.shape
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# fit model
model.fit(X=data, y=y)
```

```python
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
```

### Prediction

```python
pred = model.predict(data)
pred
```

```python
plt.scatter(X, y)
plt.plot(X, pred, color='red')
plt.show()
```

# Mutivariate Regression
### Sample data

```python
bias = 1
beta = np.array([2,3,4,5]).reshape(4, 1)
noise = np.random.randn(100, 1)
X = np.random.randn(100, 4)
y = bias + X.dot(beta)
y_with_noise = y + noise
```

```python
X[:10]
```

```python
y[:10]
```

### Regression

```python
model = LinearRegression()
model.fit(X, y_with_noise)
```

```python
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
```

# Polynomial Regression
### Dummy data

```python
bias = 1
beta = np.array([2,3]).reshape(2, 1)
noise = np.random.randn(100, 1)
X = np.random.randn(100, 1)
X_poly = np.hstack([X, X**2])
```

```python
X_poly[:10]
```

```python
y = bias + X_poly.dot(beta)
y_with_noise = y + noise
```

```python
plt.scatter(X, y_with_noise)
```

### Regression

```python
model = LinearRegression()
model.fit(X_poly, y_with_noise)
```

```python
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
```

### Prediction and Plot

```python
pred = model.predict(X_poly)
plt.scatter(X, pred, color='red')
plt.scatter(X, y_with_noise, alpha=0.3)
plt.show()

```
